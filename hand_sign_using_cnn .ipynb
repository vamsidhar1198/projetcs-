```python
import glob
import os 
import numpy as np
import cv2
import mediapipe
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import keras 
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten
from keras.utils import np_utils
```

# getting all the images from train and test folders

```python
label_list = [chr(i) for i in range(65,90)]
label_list.remove('J')
train_data = []
train_labels = []
test_data = []
test_labels = []
for i in label_list:
    hand_sign_images = glob.glob(r'/Users/vamsi/Downloads/archive/Train/'+str(i)+'/*.*')
    for j in hand_sign_images:
        img=cv2.imread(j)
        
        train_data.append(img)
        train_labels.append(i)
for i in label_list:
    hand_sign_images = glob.glob(r'/Users/vamsi/Downloads/archive/Test/'+str(i)+'/*.*')
    for j in hand_sign_images:
        img=cv2.imread(j)
        test_data.append(img)
        test_labels.append(i)  
```

```python
train_data = np.array(train_data)
train_labels = np.array(train_labels)
test_data = np.array(test_data)
test_labels = np.array(test_labels)
```

```python
train_label = pd.DataFrame(train_labels)
test_label = pd.DataFrame(test_labels)
```

```python
for i in train_label.columns:
    if train_label[i].dtype=='object':
        encoded = pd.get_dummies(train_label[i])
        train_label = train_label.join(encoded).drop(i, axis=1)
for i in test_label.columns:
    if test_label[i].dtype=='object':
        encoded = pd.get_dummies(test_label[i])
        test_label = test_label.join(encoded).drop(i, axis=1)
```

```python
model = Sequential()
model.add(Conv2D(20,(5,5),padding='valid',activation='relu',input_shape=(28,28,3))) #convolutional layer
model.add(MaxPooling2D(pool_size=(2,2))) #pooling layer 
model.add(Conv2D(10,(5,5),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten()) #convert image to a vector 

model.add(Dense(120,activation='relu')) # 120 neruons 
model.add(Dense(24,activation='softmax')) # output layer
```

```python
model.summary()
```

```python
train_data
```

```python
model.compile(optimizer="Adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(train_data,train_label,batch_size=10,epochs=100,verbose=1,validation_split=0.2)
```

```python
model.evaluate(test_data,test_label)
```

```python
label = 'HI'
labels={0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'k',10:'L',11:'M',12:'N',13:'O',14:'P',15:'Q',16:'R',17:'S',18:'T',19:'U',20:'V',21:'W',22:'X',23:'Y'}
drawingModule = mediapipe.solutions.drawing_utils
handsModule = mediapipe.solutions.hands
capture = cv2.VideoCapture(0)
x=[]
y=[]
with handsModule.Hands(static_image_mode=False, min_detection_confidence=0.7, min_tracking_confidence=0.7,max_num_hands=2) as hands:
        z=0
        while (True):
            ret, frame = capture.read()
            results = hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            if results.multi_hand_landmarks:
                for handLandmarks in results.multi_hand_landmarks:
                    drawingModule.draw_landmarks(frame, handLandmarks, handsModule.HAND_CONNECTIONS)
                    for i in range(0,21):
                        x.append(results.multi_hand_landmarks[0].landmark[i].x)
                        y.append(results.multi_hand_landmarks[0].landmark[i].y)
                    x1=min(x)
                    x2=max(x)
                    y1=min(y)
                    y2=max(y)
                    x.clear()
                    y.clear()
                    h, w, c = frame.shape
                    x11= int(x1 *w)
                    y11 = int(y1*h)
                    x21 = int(x2 *w)
                    y21 = int(y2*h)
                    cv2.rectangle(frame,(x11-30,y11-30),(x21+30,y21+30),(255,0,255),1)
                    image1 = frame[y11-30:y21+30,x11-30:x21+30]
                    if cv2.waitKey(1) == ord('t'):
                        #cv2.imwrite(r"/Volumes/study/face_mask_data_kaggel/{}.jpg".format(z),image1)
                        #z=z+1
                        image1 = image1/255
                        image2 = cv2.resize(image1,(28,28))
                        image21 = image2.reshape(1,image2.shape[0],image2.shape[1],image2.shape[2])
                        #cv2.imwrite(r"/Volumes/study/face_mask_data_kaggel/0.jpg",image2)
                        pre = model.predict(image21)
                        label = labels[np.argmax(pre)]
                        print("the label is",label)
            cv2.putText(frame,label,(100,100),cv2.FONT_HERSHEY_SIMPLEX,1, (209, 80, 0, 255),3) 
            cv2.imshow('Test hand', frame)
            if cv2.waitKey(1) == ord('q'):
                break
cv2.destroyAllWindows()
capture.release()
```
