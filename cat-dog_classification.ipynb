```python
import cv2
import os
import numpy as np
import pandas as pd 
import warnings 
warnings.filterwarnings('ignore')
```

```python
os.chdir(r"E:\vamsi\ml assignment 3\train")
files = os.listdir()
```

getting all th files names from the file train

```python
folder_names = ['cat','dog']
for i in folder_names:
    os.mkdir(r"E:\\vamsi\\ml assignment 3\\"+i)
```

creating two new folders named cat and dog

```python
for i in files:
    if i[0] == 'c':
        src_path = os.path.join(r"E:\vamsi\ml assignment 3\train", i)
        dst_path = os.path.join(r"E:\vamsi\ml assignment 3\cat", i)
        os.rename(src_path, dst_path)
    else:
        src_path = os.path.join(r"E:\vamsi\ml assignment 3\train", i)
        dst_path = os.path.join(r"E:\vamsi\ml assignment 3\dog", i)
        os.rename(src_path, dst_path)
```

moving the cat and dog images from train folder to respective cat and dog folders

```python
os.chdir(r"E:\vamsi\ml assignment 3\cat")
cat_files = os.listdir()
grey_array = []
for j in cat_files:
    grey_image = cv2.imread("E:\\vamsi\\ml assignment 3\\cat\\"+j, 0) # color to greyscale
    grey_image = cv2.resize(grey_image,(100,100)) # resizing image
    grey_array.append(grey_image.flatten().tolist())
```

converting color image to grey scale , resizing it to (100,100) then converting it to vector and appending it to dataframe with vector and label

```python
cat = pd.DataFrame(grey_array)
```

```python
cat['label']='cat'
cat
```

```python
os.chdir(r"E:\vamsi\ml assignment 3\dog")
dog_files = os.listdir()
gery_array = []
for k in dog_files:
    grey_image = cv2.imread("E:\\vamsi\\ml assignment 3\\dog\\"+k, 0)
    grey_image = cv2.resize(grey_image,(100,100))
    gery_array.append(grey_image.flatten().tolist())
```

```python
dog = pd.DataFrame(gery_array)
dog['label']='dog'
dog
```

```python
cat_dog = pd.concat([cat,dog])
```

merging both dataframes (cat,dog) into one dataframe (cat_dog)
